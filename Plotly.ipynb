```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
import plotly.express as px
import plotly.graph_objects as go
from plotly import data
```

```python
plt.style.use('ggplot')
```

# Line Plots

```python
df_stocks = data.stocks()
df_stocks.head(2)
```

```python
px.line(df_stocks,x='date',y=['GOOG','AMZN'],title='Google vs Amazon')
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df_stocks['date'],y=df_stocks['GOOG']))
```

# Bar Plots

```python
df_us = data.gapminder().query("country == 'United States'")

fig = px.bar(df_us,x='year',y='gdpPercap',color='year',text='gdpPercap')
fig.update_traces(texttemplate="%{text:.2s}")
fig.update_layout(xaxis_tickangle=45)
```

# Scatter Plots

```python
df_iris = data.iris()
px.scatter(df_iris,x='sepal_length',y='sepal_width',color='sepal_length',hover_data=['sepal_width'])
```

```python
fig = go.Figure()
fig.add_trace(
    go.Scatter(x=df_iris['sepal_length'],y=df_iris['sepal_width'],mode='markers',text=df_iris['sepal_width'])
)
```

# Pie Charts

```python
df_asia = data.gapminder().query('year==2007')

px.pie(df_asia,names='country',values='pop',color_discrete_sequence=px.colors.sequential.RdBu)
```

```python
fig = go.Figure(data=[
    go.Pie(names=df_asia['country'],values=df_asia['pop'])
])
```

# Histogram

```python
import numpy as np

dice1 = np.random.randint(1,7,5000)
dice2 = np.random.randint(1,7,5000)
dice = dice1 + dice2

fig = px.histogram(dice,nbins=11,labels={'value':'Dice'})
fig.update_layout(xaxis_title_text="X title ",yaxis_title_text="Y title ",bargap=0.2,showlegend=False)
```

# Box Plots

```python
df_tips = data.tips()
px.box(df_tips,x='sex',y='tip',color='sex')
```

```python
fig = go.Figure()
fig.add_trace(
    go.Box(x=df_tips['sex'],y=df_tips['tip'],marker_color='red',boxmean='sd')
)
```

# Violin Plots

```python
px.violin(df_tips,x='sex',y='tip',box=True,color='sex')
```

```python
fig = go.Figure()
fig.add_trace(
    go.Violin(x=df_tips['sex'],y=df_tips['tip'],legendgroup='yes')
)
```
